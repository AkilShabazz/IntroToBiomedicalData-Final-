#### Preprocessing Data and Show Table

```python
import pandas as pd

# import 12 lead ECG dataset
df = pd.read_csv('MUSE_20180111_155115_19000.csv', header=None)

df.head()
```

#### Labelling Columns

```python
#list to be used as column names
new_cols_list = ['I', 'II', 'III', 'aVR', 'aVL', 'aVF', 'V1', 'V2', 'V3', 'V4', 'V5', 'V6']

# renames cols with a loop
for i in range(0,12):
    df.columns = new_cols_list

df.head()
```

```python
df.plot(subplots=True, layout=(6,2),figsize=(15,15))
```
